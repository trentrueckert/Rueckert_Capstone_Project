# Exploratory Data Analysis (EDA)
## Install Dependencies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load the Data

```python
# Load the data and conduct descriptive statistics
df = pd.read_csv("Data/Cleaned_OddsData.csv")

print(df.head)
print(df.shape)

print(df.describe())
print(df.info())
```

## Check for Missing/Null Values and Duplicates

```python
print('Missing Values:')
print(df.isnull().sum(), '\n')

print(f"Duplicate Rows: {df.duplicated().sum()}")
```

## Numerical Column Distribution Plots

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
    plt.show()
```

## Distribution of Categorical Columns

```python
cat_cols = df.select_dtypes(include='object').columns.drop('date')

for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f"Count of {col}")
    plt.xticks(rotation=90)
    plt.show()
```

## Bivariate Analysis

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap')
plt.show()
```
